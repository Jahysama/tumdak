```python
import json
from tqdm import tqdm
import numpy as np
import pandas as pd

user_dict = {}
df = pd.read_json('/content/telegram/blagoveshensk_chat.json')
with open('/content/telegram/blagoveshensk_chat.json', 'r') as chat:
  usr_list = []
  chat_log = json.load(chat)
  for message in tqdm(chat_log):
    if not message["user_id"] in usr_list:
      usr_list.append(message["user_id"])
  dim = len(usr_list)
  emb = np.zeros((dim,dim), dtype=np.int8)


df_replies = df[df['reply_message_id'].notna()]
for index, line in tqdm(df_replies.iterrows()):
  if line['reply_message_id']:
    user_reply = df.loc[df['message_id'] == line['reply_message_id']]["user_id"].tolist()
    if user_reply:
      user_reply = user_reply[0]
      emb[usr_list.index(user_reply)][usr_list.index(line["user_id"])] += 1
```

```python
import seaborn as sns
from matplotlib import pyplot

a4_dims = (100, 120)
fig, ax = pyplot.subplots(figsize=a4_dims)
sns.heatmap(emb,ax=ax)
```
